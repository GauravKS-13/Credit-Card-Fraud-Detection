# Feature Engineering

```python
# Dependent and independent feature
x = dataset.iloc[:,:-1]
y = dataset.iloc[:,-1]
print(x.shape,y.shape)
```

#### we will balance our dataset usig Under sampling

```python
# we are importing these libraries to work well when performing under sampling
import six
import sys
sys.modules['sklearn.externals.six'] = six
import joblib
sys.modules['sklearn.externals.joblib'] = joblib
```

```python
from imblearn.under_sampling import RandomUnderSampler
ros = RandomUnderSampler(random_state=0)
x_resampled,y_resampled = ros.fit_sample(x, y)
```

```python
print(x_resampled.shape,y_resampled.shape)
```

```python
X = pd.DataFrame(x_resampled,columns = x.columns)
X.head()
```

```python
Y = pd.Series(y_resampled,name = 'Class') 
```

```python
resampled_dataset = pd.concat([X,Y],axis=1)
resampled_dataset.head()
```

### Train Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,Y,test_size =0.20,random_state =0)
print(x_train.shape,x_test.shape)
```

```python
#  for Feature Selection

from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

### Apply the feature selection
# first specify the lasso regression model than
# chooose the suitable value of alpha ( equivalent of penlety)
# higher the alpha value less featue will be selected

# then we use the SelectFromModle from sklearn which will select the features which have the coefficient are non zero

feature_sel_model = SelectFromModel(Lasso(alpha = 0.005, random_state = 0)) # Remember to set the seed and random_state same will use in test data
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
```

Feature those have True value they are selected and have non-zero coefficient value and features which has False are not Selected and have zero coefficient value

```python
selected_features = X.columns[(feature_sel_model.get_support())]
print(selected_features,"\n")

print("Total Selected Features are {}".format(len(selected_features)))
```

```python


x_train = x_train[['V1', 'V2', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V13',
                  'V14', 'V15', 'V18', 'V19', 'V20', 'V21', 'V22', 'V24', 'V26', 'V28','Amount']]

x_test = x_test[['V1', 'V2', 'V4', 'V5', 'V6', 'V7', 'V8', 'V9', 'V10', 'V11', 'V13',
                'V14', 'V15', 'V18', 'V19', 'V20', 'V21', 'V22', 'V24', 'V26', 'V28', 'Amount']]
x_train.shape,x_test.shape
```

# Model Building

```python
# We are using Random Forest classifier for our model building

from sklearn.ensemble import RandomForestClassifier
rf_cls = RandomForestClassifier(random_state =0)
rf_cls.fit(x_train,y_train)
```

```python
y_predict = rf_cls.predict(x_test)
y_predict[0:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score
confusion_matrix = confusion_matrix(y_test,y_predict)
classification_report = classification_report(y_test,y_predict)
print(confusion_matrix)
sns.heatmap(confusion_matrix,annot = True)
```

```python
print(classification_report)
```

```python
f1_score = f1_score(y_test,y_predict)
f1_score
```

A good F1 score means that you have low false positives and low false negatives, so you're correctly identifying real threats and you are not disturbed by false alarms. An F1 score is considered perfect when it's 1 , while the model is a total failure when it's 0 .

### Cross validation on the Training data 

```python
# After the class balance we hace less data records so we will use corss validation to see how model is performing.
# It will give us better idea how will our model will behave on unseen data.

rom sklearn.model_selection import cross_val_score
rf_cls_cv = RandomForestClassifier()
cross_val_score = cross_val_score(rf_cls_cv,x_train,y_train,cv = 5,scoring = 'f1')
print(cross_val_score)
print("F1 Score is ",np.mean(cross_val_score))
```

A good F1 score means that you have low false positives and low false negatives, so you're correctly identifying real threats and you are not disturbed by false alarms. An F1 score is considered perfect when it's 1 , while the model is a total failure when it's 0 .

## Improving our model by Hyperparameter Tunning

```python
from sklearn.model_selection import RandomizedSearchCV
rf_clf_rsv = RandomForestClassifier(random_state = 0)
parameters = { 'n_estimators':  [100,150,200,250,300,350,400],
               'max_depth' : [3,4,5,6,7,8,9,10],
               'min_samples_split' : [5,7,9,10,11,12,14],
               'max_features' : ["auto", "sqrt", "log2"] }
```

```python
random_scv = RandomizedSearchCV(estimator = rf_clf_rsv, param_distributions = parameters,n_iter=10, scoring = 'f1', n_jobs=-1,random_state=0)
search = random_scv.fit(x_train,y_train)
search.best_estimator_
```

```python
# we can get best hyperparameters valuse for our model   
search.best_params_
```

```python
# Making prediction with our model with new hpyerparametes
y_pred = search.predict(x_test[selected_features])
y_pred[0:5]
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score,f1_score
confusion_matrix = confusion_matrix(y_test,y_pred)
classification_report = classification_report(y_test,y_pred)
print(confusion_matrix)
sns.heatmap(confusion_matrix,annot= True)
```

```python
print(classification_report)
```

```python
f1_score = f1_score(y_test,y_pred)
f1_score
```

### We have got the 0.9604  F1 Score on the test data. We can see that our model is performining very well on the new data(test data) after the hyperparameter tuning.
